```python
def rep(x) :
    
    x = str(x)
    
    if '역' in x :
        return x
    
    return x + '역'


data_1['STATION_NM'] = data_1['STATION_NM'].apply(lambda x : rep(x))
```

```python
import re

def rep_2(x) :
    
    x = str(x)
    
    if '(' in x :
        return re.sub(r"[(]\w*[)]", '', x)
    
    return x
```

```python
data_1['STATION_NM'] = data_1['STATION_NM'].apply(lambda x : rep_2(x))
```

```python
print(set(data_main['역사명']) - set(data_1['STATION_NM']))
```

```python
data_1[data_1['STATION_NM'].str.contains('동대문')]
```

```python
def rep_3(x) :
    
    x = str(x)
    
    if x in ['역삼', '동대문역사문화공원', '역촌'] :
        return x + '역'
    
    elif x == '남한산성입구(성남법원.검찰청)역' :
        return '남한산성입구역'
    
    elif x == '교대(법원.검찰청)역' :
        return '교대역'
    
    return x
```

```python
data_1['STATION_NM'] = data_1['STATION_NM'].apply(lambda x : rep_3(x))
```

```python
print(set(data_main['역사명']) - set(data_1['STATION_NM']))
```

```python
import re

def re_dot(x) :
    
    x = str(x)
    
    if '·' in x :
        return re.sub(r"[·]", '', x)
    
    return x
```

```python
data_2['행정동'] = data_2['행정동'].apply(lambda x : re_dot(x))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca_data = pca.fit_transform(pca_choice)
print(pca_data.shape)
```
